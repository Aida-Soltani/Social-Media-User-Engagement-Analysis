```python
# -Convert the given string status_published to a datetime object
# -Extract year,month, day and hour of the published post
```

```python
import datetime
```

```python
df['Date_published']= df['status_published'].apply(lambda x:datetime.datetime.strptime(x, '%m/%d/%Y %H:%M'))
df['Year_published'] =df['Date_published'].apply(lambda x: x.year)
df['Month_publish'] = df['Date_published'].apply(lambda x: x.month)
df['Day_publish'] = df['Date_published'].apply(lambda x: x.day)
df['Hour'] = df['Date_published'].apply(lambda x: x.hour)
```

```python
# ...........Support Vector Regression............#
```

```python
X = data[['Day_publish','Hour']]
y = data['Engement_Score']
```

```python
#X = data[['Year_published']]
#y = data['num_shares']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train = X_train.values
y_train = y_train.values
X_test = X_test.values
y_test = y_test.values
```

```python
# X_train = X_train.reshape(1, -1)
y_train= y_train.reshape(-1, 1)
# X_test = X_test.reshape(1, -1)
y_test= y_test.reshape(-1, 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X_train)
y = sc_y.fit_transform(y_train)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X, y)
```

```python
y_pred = regressor.predict(X_test)
y_pred = sc_y.inverse_transform(y_pred) 
```

```python
from sklearn import metrics
```

```python
MAE = metrics.mean_absolute_error (y_test, y_pred)
MSE = metrics.mean_squared_error (y_test, y_pred)
RMSE = np.sqrt(MSE)
```
